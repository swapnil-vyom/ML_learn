Imagine the word "cats" appears at position 1 in:
"Cats rule the world"
"All animals except cats are here"
The model needs to treat "cats" differently depending on where it appears.
The sine function helps by:
Assigning a unique pattern to each position
Ensuring nearby positions have similar patterns
Ensuring far positions have very different patterns

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
d_model = 512
pos = 10  # Fixed position
i_values = list(range(0, d_model, 2))  # Even dimensions

# Calculate the denominator (frequency term)
denominators = [10000 ** ((2 * i) / d_model) for i in i_values]
values = [pos / denom for denom in denominators]
sine_values = [np.cos(v) for v in values]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(i_values, sine_values, label=f"sin(pos / (10000^(2i/d_model))) at pos={pos}")
plt.title("Variation of Sine Positional Encoding with Dimension Index (i)")
plt.xlabel("Dimension index (i)")
plt.ylabel("Sine value")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
d_model = 512
pos = 10  # Fixed position
i_values = list(range(1, d_model, 2))  # Odd dimensions

# Calculate the denominator (frequency term)
denominators = [10000 ** ((2 * (i + 1)) / d_model) for i in i_values]
values = [pos / denom for denom in denominators]
cosine_values = [np.cos(v) for v in values]

# Plot
plt.figure(figsize=(12, 6))
plt.plot(i_values, cosine_values, label=f"cos(pos / (10000^(2i/d_model))) at pos={pos}", color='orange')
plt.title("Variation of Cosine Positional Encoding with Dimension Index (i)")
plt.xlabel("Dimension index (i)")
plt.ylabel("Cosine value")
plt.grid(True)
plt.legend()
plt.show()
```
